```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("CC GENERAL.csv")
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
mean_value=df['CREDIT_LIMIT'].mean()
df['CREDIT_LIMIT']=df['CREDIT_LIMIT'].fillna(mean_value)
mean_value=df['MINIMUM_PAYMENTS'].mean()
df['MINIMUM_PAYMENTS']=df['MINIMUM_PAYMENTS'].fillna(mean_value)
```

```python
df.drop("CUST_ID", axis = 1, inplace = True)
```

```python
plt.figure(figsize=(10,6))
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(df)
```

```python
best_cols = ["BALANCE", "PURCHASES", "CASH_ADVANCE","CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS"]
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
best_vals = df[best_cols].iloc[ :, 1:].values
y_hc = hc.fit_predict( best_vals )

df["cluster"] = y_hc
best_cols.append("cluster")
sns.pairplot( df[ best_cols ], hue="cluster")
```

# K-means

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'random', random_state = 101)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'random', random_state = 101)
y_kmeans = kmeans.fit_predict(df)
```

```python
best_cols = ["BALANCE", "PURCHASES", "CASH_ADVANCE","CREDIT_LIMIT", "PAYMENTS", "MINIMUM_PAYMENTS"]
kmeans = KMeans(n_clusters=4, init="random", max_iter=300) 
best_vals = df[best_cols].iloc[ :, 1:].values
y_pred = kmeans.fit_predict( best_vals )

df["cluster"] = y_pred
best_cols.append("cluster")
sns.pairplot( df[ best_cols ], hue="cluster")
```

```python
from sklearn import metrics
metrics.silhouette_score(df, kmeans.labels_)
```

```python
metrics.silhouette_score(df, hc.labels_)
```

As you can see in the above figure clusters (for k means and hierarchical clusetring) are not well apart. The inter cluster distance is almost negligible 0.4 < 1
