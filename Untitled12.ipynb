```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import re
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(data).transform(data)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
count = df.loc[:,:].sum()
df1 = count.sort_values(0, ascending = False)
df1 = df1.to_frame()
df1 = df1.reset_index()
df1 = df1.rename(columns = {"index": "items",0: "count"})
```

```python
ax = df1.plot.bar(x = "items", y = "count")
plt.title("Top Selling items")
```

```python
frequent_itemsets=apriori(df, min_support=0.5, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.6) 
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
dataset.head()
```

```python
dataset.fillna(0,inplace=True)

transactions = []

for i in range(0,len(dataset)):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20) if str(dataset.values[i,j])!='0'])
```

```python
te=TransactionEncoder()
te_ary=te.fit(transactions).transform(transactions)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
count = df.loc[:,:].sum()
df1 = count.sort_values(0, ascending = False).head(20) # Top 20 Selling Items
df1 = df1.to_frame()
df1 = df1.reset_index()
df1 = df1.rename(columns = {"index": "items",0: "count"})
```

```python
ax = df1.plot.bar(x = "items", y = "count")
plt.title("Top Selling items")
```

```python
frequent_itemsets=apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3) 
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```
